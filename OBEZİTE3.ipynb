```python
# Kütüphanelerin çağırılması
import pandas as pd
```

```python
# Eğer veri dosyası başka bir klasörün içindeyse Klasörün adını da belirtmemiz gerekiyor.
obezite = pd.read_csv("https://raw.githubusercontent.com/nurefsanoztekin/data_k/main/OBEZ%C4%B0TE-DATA.csv")
obezite
```

```python
## head(x): veri setindeki ilk x adet veriyi gösterir.

obezite.head(4)
```

```python
## tail(x): veri setindeki son x adet veriyi gösterir.

obezite.tail(7)
```

```python
## sample(x): veri setinden rasgele olarak x adet veriyi örnekler.

obezite.sample(10)
```

```python
# shape: verinin boyutlarını gösterir. (x:satır, y:sütun)
obezite.shape
```

```python
# columns: sütun isimlerini verir.
obezite.columns
```

```python
# info():  veri hakkında genel bilgi verir. 
# (Kaç sütun olduğu, sütun isimleri, her sütunda kaç örnek olduğu, eksik örnek olup olmadığı ve veri tipleri)
obezite.info()
```

```python
# describe():  verinin istatistiksel bir özetini sunar.
# Count: örnek sayısı, mean: ortalama, std: standart sapma, min: minimim değer
# 25%: Q1, %50: Q2 veya median, 75%: Q3, max: maksimum değer

obezite.describe()
```

```python
## For Döngüsü ile Tüm sütunların değerlerini ve sayılarını yazdırma
for i in (obezite.columns):
    print(i)
    print(obezite[i].value_counts())
    print("\n")
```
